```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
```

```python
COVID = pd.read_csv('Covid19_case_summary.csv')     
COVID = COVID.set_index('description')
COVID = COVID.drop(index=['State Data','Case Rates Per 100,000 People in Colorado by County','Cases of COVID-19 Reported in Colorado by Age Group, Hospitalization, and Outcome','Colorado Case Counts by County','COVID-19 in Colorado by Sex','Case Status for Cases & Deaths','COVID-19 in Colorado by Age Group','COVID-19 in Colorado by Race & Ethnicity'])
COVID.head(20)
```

```python
Covid2=COVID.pivot_table(index=['attribute','metric'],columns='description',values='value')
Covid2.head(100)

Covid2.plot(x='attribute',y='Cases of COVID-19 in Colorado by Date Reported to the State')
```

```python
plt.rcParams['figure.figsize'] = [40/2.54, 20/2.54]
fig, axs = plt.subplots(nrows=2, ncols=1,facecolor='white')

######## By Date Reported ###########
#Plot 7-day avg line
N=len(COVID)
COVID.plot(x="Date",y="7DayAvg",linewidth=2,label='7 day avg',color='k',ax=axs[0])

#Fill with colors by slope
cmap = mpl.cm.RdBu.reversed()
normalize = mpl.colors.Normalize(vmin=-80, vmax=80)
for i in range(1,N):
    axs[0].fill_between(x=COVID["Date"][i-1:i+1],y1=COVID["7DayAvg"][i-1:i+1],y2=0, color=cmap(normalize(COVID["d7Day"][i])))

#Plot daily points
COVID.plot(x="Date",y="dCases",label="Daily value",kind='scatter',ax=axs[0],color='dimgrey')

#Make plot look better
xTicks=np.arange(12, N, 7)
axs[0].set(xlabel='Date',ylabel='New Cases Per Day (Reported)',xticks=xTicks,xticklabels=COVID["Date"][xTicks],xlim=(7, N-1),ylim=(0, 600))
#ax.set_yscale('log')
axs[0].grid()

#show colorbar
sm = plt.cm.ScalarMappable(cmap=cmap, norm=normalize)
sm.set_array([])
plt.colorbar(sm,ax=axs[0],orientation='vertical',label='$\Delta$ 7-day avg new cases')

############# By Onset Date #############
N=len(Onset)
Onset.plot(x="Date",y="7DayAvg",linewidth=2,label='7 day avg',color='k',ax=axs[1])

cmap = mpl.cm.RdBu.reversed()
normalize = mpl.colors.Normalize(vmin=-30, vmax=30)

for i in range(1,N):
    axs[1].fill_between(x=Onset["Date"][i-1:i+1],y1=Onset["7DayAvg"][i-1:i+1],y2=0, color=cmap(normalize(Onset["d7Day"][i])))
    
Onset.plot(x="Date",y="Cases",label="Daily value",kind='scatter',ax=axs[1],color='dimgrey')

xTicks=np.arange(8, N, 7)
axs[1].set(xlabel='Date',ylabel='New Cases Per Day (by Onset)',xticks=xTicks,xticklabels=Onset["Date"][xTicks],xlim=(7, N-1),ylim=(0, 600))
axs[1].grid()
plt.xticks(rotation=30)


sm = plt.cm.ScalarMappable(cmap=cmap, norm=normalize)
sm.set_array([])
plt.colorbar(sm,ax=axs[1],orientation='vertical',label='$\Delta$ new cases')
```
